**REDUCING RESOLUTION OF AN IMAGE**

```python
import sys
import numpy as np
from PIL import Image
```

Get Image

```python
rose = Image.open("Rose1024.tif")
rose
```

Check the size of array and its values

```python
rose_arr = np.array(rose)
print(rose_arr.shape)

print([i for i in rose_arr[0]])
#with np.printoptions(threshold=sys.maxsize):
#    print(rose_arr)
```

Create Functions to Reduce Resolution and Display the Images

```python
# Function that decreases resolution by specified times.
def reduce_resolution(arr, reduce_by):
  current_res = arr.shape[0]
  new_image = []

  # Every 1 of 2 element were removed from every rows.
  for index in arr[0::reduce_by]:
    line = []
    for byte in index[0::reduce_by]:
      line.append(byte)
    new_image.append(line)

  return new_image
print()

# Function to display the picture and its shape.
def display_image(arr):
  reduced_image = Image.fromarray(arr)
  print([i for i in arr[0]])  # First row were printed to check arrays.
  new_size = (300,300)                           # To see the difference on the 
  reduced_image = reduced_image.resize(new_size) # details, images were sized.
  reduced_image.show()
  print(arr.shape)
  print("*"*75)

# Reducing resolutions.
rose512_arr = np.array(reduce_resolution(rose_arr, 2))
rose256_arr = np.array(reduce_resolution(rose512_arr, 2))
rose128_arr = np.array(reduce_resolution(rose256_arr, 2))
rose64_arr = np.array(reduce_resolution(rose128_arr, 2))
rose32_arr = np.array(reduce_resolution(rose64_arr, 2))

# Display images with sizes.
display_image(rose_arr)
display_image(rose512_arr)
display_image(rose256_arr)
display_image(rose128_arr)
display_image(rose64_arr)
display_image(rose32_arr)

```

**REDUCING THE BITS OF AN IMAGE (FROM 8 BITS TO 4)**

```python
ctskull = Image.open("Ctskull-256.tif")

width = ctskull.width
height = ctskull.height

print(width, height)
```

```python
img_bytes = ctskull.tobytes()

byte_to_int = np.frombuffer(img_bytes, dtype=np.uint8)  # 169048 bytes total

print(byte_to_int[0])
bits_8to4 = (byte_to_int/255)*127   # Convert 8 bits to 4 bits.

bits_8to4 = bits_8to4.astype(int)
print(bits_8to4[0::5000])   # Just to check mapping of the bits
```

```python
image_arr_4bits = bits_8to4.tobytes()
print(img_bytes[0])
skull_4bits = Image.frombytes('L',(width,height), image_arr_4bits)
skull_4bits.show()
ctskull.show()

print(len(image_arr_4bits))   # 1352384
print(type(image_arr_4bits))
print(type(bits_8to4[0]))
```
